# Ejemplos de ciclo while

```python
# 1. Ejemplo de while interactivo con operadores relacionales
# Este tipo de estructura ejecuta un bloque de codigo mientras la condicion se cumple

numero = int(input("Ingresa un numero: ")) # Solicita un numero al usuario

contador = 1 # Asigna el valor 1 a la variable 'contador'

while contador <= numero: # Evalúa si 'contador' es menor o igual al numero ingresado
    print("Contador:", contador) # Se ejecuta mientras la condición anterior es True
    contador += 1 # Incrementa el contador en 1
```

```python
# 2. Ejemplo de while interactivo con operadores logicos y relacionales
# Se utiliza para repetir acciones hasta cumplir una condición

numero = int(input("Ingresa un numero par: ")) # Define un número entero

while numero % 2 != 0: # Usa el operador aritmético módulo (%) y el relacional (!=)
    print("El numero no es par.") # Se ejecuta si el numero es impar
    numero = int(input("Ingresa otro numero: ")) # Solicita otro numero

print("El numero es par.") # Se ejecuta cuando la condición es False
```

```python
# 3. Ejemplo de while utilizando operadores lógicos, relacionales y aritméticos
# Permite validar múltiples condiciones

nota = int(input("Ingresa tu calificacion: ")) # Variable con una calificación
asistencia = int(input("Ingresa tu asistencia: ")) # Variable con porcentaje de asistencia

while nota < 0 or nota > 100: # Evalúa condiciones alternativas con 'or'
    nota = int(input("Calificacion invalida, ingresa otra: "))

if nota >= 90 and asistencia >= 80: # Evalúa dos condiciones con el operador lógico 'and'
    print("Excelente desempeño.") # Ejecuta si ambas condiciones son True
elif nota >= 70 or asistencia >= 95:
    print("Aprobado por notas o gran asistencia.") # Ejecuta si al menos una es True
else:
    print("Debes esforzarte más.") # Ejecuta si ninguna condición se cumplió
```

```python
# 4. Ejemplo combinado (Cálculo de descuento)
# Este ejemplo muestra cómo se integran operaciones aritméticas dentro de la lógica de repetición

precio_unitario = 100 # Precio base
cantidad = int(input("Ingresa la cantidad comprada: ")) # Cantidad comprada

while cantidad <= 0: # Comprueba que la cantidad sea válida
    cantidad = int(input("Cantidad invalida, ingresa otra: "))

if cantidad > 10: # Comprueba si hay más de 10 artículos
    total = (precio_unitario * cantidad) * 0.90 # Aplica 10% de descuento
    print(f"Total con descuento: {total}") # Muestra el resultado calculado
else:
    total = precio_unitario * cantidad # Calcula el total sin descuento
    print(f"Total sin descuento: {total}") # Muestra el resultado
```

```python
# 5. Ejemplo interactivo de while con contraseña
# Este programa solicita la contraseña hasta que sea correcta

contrasena = "python123" # Contraseña correcta

entrada = input("Ingresa la contraseña: ") # Solicita la contraseña

while entrada != contrasena: # Evalúa si la contraseña es incorrecta
    print("Contraseña incorrecta") # Mensaje de error
    entrada = input("Ingresa nuevamente la contraseña: ") # Solicita otra vez la contraseña

print("Acceso permitido") # Se ejecuta cuando la contraseña coincide
```
